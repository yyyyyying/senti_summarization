```python
# encoding=utf-8
import jieba
from gensim.corpora import Dictionary
from gensim.models import LdaModel
from gensim.models import ldaseqmodel
stopwords = [line.strip() for line in open('./chinese_stopword.txt','r').readlines()]
```

```python
train=[]
time_slice=[]
f= open('../dtm/jg_wb','r')
info=eval(f.readline().strip()) 

for i in info:
    time_slice.append(i[1][0])

line=eval(f.readline().strip()) 

for i in f.readlines():
    i=eval(i.strip())
    train.append(i[3])
```

```python
jieba.add_word(u'江歌') 
jieba.add_word(u'江歌案') 
jieba.add_word(u'江秋莲') 
```

```python
train_set = []
for line in train:
    line = list(jieba.cut(line))
    train_set.append([ w for w in line if w not in stopwords and len(w)>1 ]) #不是听用词 且 长度>1（去除单个字）
```

```python
dictionary = Dictionary(train_set)
corpus = [ dictionary.doc2bow(text) for text in train_set]
```

```python
#主题数设置问题
ldaseq = ldaseqmodel.LdaSeqModel(corpus=corpus, id2word=dictionary, time_slice=time_slice, num_topics=5)
```

```python
#某个时间的主题分布
print('ldaseq_topic----------\n',ldaseq.print_topics(time=0))
```

```python
#某个主题各个时间的分布
print('ldaseq_topic_times----------\n',ldaseq.print_topic_times(topic=0))
```
